# 촬영한 비디오를 이미지로 변환

```python
__author__ = 'Seran'

import cv2

# 영상의 이미지를 연속적으로 켭쳐할 수 있게 하는 class
# vidcap = cv2.VideoCapture('test1.mp4')

count = 0

# fps = int(vidcap.get(cv2.CAP_PROP_FPS))
```

```python
for i in range(3):
    vidcap = cv2.VideoCapture(f'test{i+1}.mp4')
    
    fps = int(vidcap.get(cv2.CAP_PROP_FPS))
    
    while(vidcap.isOpened()):
        # read()는 grab()와 retrieve() 두 함수를 한 함수로 불러움
        # 두 함수를 동시에 불러오는 이유는 프레임이 존재하지 않을 때
        # grab() 함수를 이용하여 return false 혹은 NULL 값을 넘겨 주기 때문
        ret, image = vidcap.read()

        if ret:
            if(int(vidcap.get(1)) % fps == 0):
                cv2.imwrite("/Users/kwonkihyeon/Desktop/공모전 대회/images/%06d.jpg" % count, image)
                print("Saved %06d.jpg" % count)
                count += 1
        else:
            break

    vidcap.release()
```
